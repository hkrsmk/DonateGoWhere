```python
# I REALLY didn't want to end up using pandas because it's clunky, but I think it's time to do so.

import pandas as pd
import json
```

```python
charities = pd.read_json("allcharityinfojsonmake.json")
charities.head()
```

```python
grouped = charities.groupby("CharityName").agg(lambda x: list(x))
grouped.head()
```

```python
# the answer was almost correct: https://stackoverflow.com/questions/51262101/delete-all-nan-values-from-list-in-pandas-dataframe

remove_nan = grouped.applymap(lambda x: [i for i in x if str(i) != "nan"])
remove_nan.head()
```

```python
allin1 = remove_nan.reset_index()
allin1.to_json("allcharityinfoallin1.json", orient = "records")
```
